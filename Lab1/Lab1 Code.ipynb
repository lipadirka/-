```python
import numpy as np
import pandas as pd
import seaborn as sns



```

```python
data = {
'Шампунь': np.random.normal(110, 3, 12),
'Порошок': np.random.normal(150, 3, 12),
'Паста зубная': np.random.normal(120, 5, 12),
'Мыло': np.random.normal(60, 32, 12),
'Ушные палочки': np.random.normal(63, 38, 12),
'Ватные диски': np.random.normal(70, 32, 12),
'Бутылка воды 1,5л': np.random.normal(45, 23, 12),
'Яблочный сок 1л': np.random.normal(85, 31, 12),
'Молоко 1л': np.random.normal(70, 23, 12),
"Хлеб": np.random.normal(45, 29, 12),
}
```

```python
df = pd.DataFrame(data)
df
```

```python
ax = sns.lineplot(df)
sns.move_legend(ax, 'upper left', bbox_to_anchor=(1, 1)) 
```

```python
p0 = df.sum() / df.shape[0]
p0
```

```python
std = ((df - p0)**2).sum() / (df.shape[0] - 1)
std = std ** (1/2)  
```

```python
x_extrapol = p0 + np.random.normal(0, std, len(p0))
pd.concat([df, pd.DataFrame([x_extrapol], columns = x_extrapol.index)]).reset_index(drop = True)
```

```python
reliability = std/p0
reliability
```

```python
cond1 = ((df - p0) < 2 * std).all()
cond1
```

```python
cond2 = p0 > 2 * std
cond2
```

```python
cond3 = (df > 0).all()
cond3
```

```python
product_color = pd.Series(dtype = 'string')
for name in df.columns:
    if (not cond3[name]):
        product_color[name] = 'Красный'
    elif (not cond1[name] and not cond2[name]):
        product_color[name] = 'Оранжевый'
    elif (not cond1[name] or not cond2[name]):
        product_color[name] = 'Жёлтый'
    else:
        product_color[name] = 'Зелёный'
product_color
```
